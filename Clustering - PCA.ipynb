```python
def numericalGrades(letter):
    if letter == 'L':
        return 3
    elif letter == 'M':
        return 2
    elif letter == 'H':
        return 1
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()

scaled = scalar.fit_transform(df[[col for col in df.columns if col != 'class' ]])
scaled_pca = scalar.fit_transform(df_pca)
scaled_spca = scalar.fit_transform(df_spca)
```

```python
y = df['class'].apply(lambda x: numericalGrades(x))
```

```python
df['class'].head()
```

```python
y.head()
```

```python
from sklearn.decomposition import PCA, IncrementalPCA

pca = PCA(n_components=4, random_state=42, svd_solver='randomized', iterated_power=10, )
pca.fit(scaled_pca, y=df['class'])
pcafeatures_train = pca.fit_transform(scaled_pca, y=df['class'])
```

```python
pd.DataFrame(pcafeatures_train).head()
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(20, 12))
plt.scatter(pcafeatures_train[:, 0], pcafeatures_train[:, 1], 
            c=y, edgecolor='none', alpha=1,
            cmap=plt.cm.get_cmap('rainbow_r', 3), s=150)
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.colorbar();
```
